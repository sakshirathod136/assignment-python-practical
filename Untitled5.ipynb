```python
list1 = [10, 20, 30, 40]

print(20 in list1)
print(50 not in list1)
```

```python
list1 = [1, 2, 3, 4, 5]

print(list1[0])     # indexing
print(list1[-1])    # negative indexing
print(list1[1:4])   # slicing
```

```python
list1 = [1, 2, 3]

list1[1] = 10       # update
list1.append(4)     # append
list1.insert(1, 20) # insert
list1.remove(3)     # delete

print(list1)
```

```python
list1 = [1, 2]
list2 = [3, 4]

print(list1 + list2)   # concatenation
print(list1 * 2)       # repetition
print(len(list1))      # length
print(max(list1))      # maximum
print(min(list1))      # minimum
```
